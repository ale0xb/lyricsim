```python
import pandas as pd

df_idx = pd.read_csv('../data/processed/filtered_detailed_results.csv')
df_corpus = pd.read_table('../data/raw/full_lyrics.tsv')
```

```python
_df = df_corpus.set_index('song_id')
df = pd.DataFrame()
df['text1']=_df.loc[df_idx.id1].text.reset_index(drop=['song_id'])
df['text2']=_df.loc[df_idx.id2].text.reset_index(drop=['song_id'])
df.reset_index(inplace=True, drop=['index'])
df['id1'] = df_idx.id1
df['id2'] = df_idx.id2
df['labels'] = df_idx.sim_rating.astype(int) - 1
```
